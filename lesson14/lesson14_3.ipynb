```python
import random
with open('names.txt',encoding='utf-8')as file:
    name_str=file.read()
    display(name_str)
names_list=name_str.split(sep='\n')
names=random.choices(names_list,k=50)
names
```

```python
import numpy as np
import pandas as pd
scores=np.random.randint(50,101,size=(50,5))
df1=pd.DataFrame(scores,columns=['國文','英文','數學','地理','探索'],index=names)
df1
```
