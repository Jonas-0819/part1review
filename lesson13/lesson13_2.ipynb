```python
#標準寫法
import matplotlib.pyplot as plt

# 1.製作figure
fig = plt.figure()

# 2.在1.的figure上製作一個axes
ax_1 = fig.add_subplot(1, 1, 1)

# 3.將圖的數據儲存在axes中
X = [0, 1, 2]
Y = [3, 4, 5]
ax_1.plot(X, Y, 'bo-.')

# 4.顯示圖
plt.show()
```

```python
import matplotlib.pyplot as plt
fig=plt.figure()
axe1=fig.add_subplot(1,3,1)
axe2=fig.add_subplot(1,3,2)
axe3=fig.add_subplot(1,3,3)

axe1.set_title('1')
axe2.set_title('2')
axe3.set_title('3')
```

```python
import matplotlib.pyplot as plt
student1:dict[str,int] = {'chinese':78, 'english':80, 'math':92, 'history':75, 'discover':85}
student2:dict[str,int] = {'chinese':90, 'english':60, 'math':85, 'history':62, 'discover':73}
subject=list(student1.keys())
score1=list(student1.values())
score2=list(student2.values())

fig=plt.figure(figsize=(10,5))
axes=fig.add_subplot(1,1,1)
axes.plot(subject,score1)
axes.plot(subject,score2)
plt.show()
plt.show()
```
